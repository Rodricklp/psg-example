# Command & Conquer
## Overview
Comanda a tus tropas y conquista, en un campo de batallas el rival que avance más rápido y con mas tropas gana. Ve tomando el control de los fuertes para conquistar el fuerte del contrincante y derrotalo.

# Reglas del Evento
* Este juego es para 2 jugadores (blanco y negro).
* Cada jugador puede tener un máximo de población igual a 6.
* Cada jugador inicia con 3 de población en el fuerte principal que se ubica al extremo del tablero.
* Un dado de 6 caras el cual definirá quien empieza el juego, el número mayor comienza.
* El primero en jugar lanza el dado y esta obligado a mover piezas cuadrícula por cuadrícula respectivamente.
* Solo si el jugador obtiene 6, puede optar por 6 movimientos o aumentar su población en 1.
* En una cuadrícula pueden estar hasta 6 piezas del mismo color.
* Si una o más fichas avanzan a una cuadrícula donde existen fichas contrarias vence el color que mas fichas posee, si igualan en número ambos ejercitos pierden las fichas.
* El jugador que lanza el dado decide si avanzar o retroceder sus fichas.
* Cuando un jugador aumenta su población, puede escoger en que fuerte iniciar sus movimientos.
* Cuando un jugador conquista el fuerte principal del contrincante gana la partida.
